6)	Create a function to find the male athlete & female Athlete who won the highest number of medals. You should pass the Summer/Winter Dataframe object to the function. Also pass the gender to the function. The function should return the name of the male/female athlete who won the highest number of medals.

```python
def calculateNumberofMedals(dataset,gender):
    copy_data=dataset.copy()       #Create a copy of dataframe
    seperated_data=copy_data[['Athlete','Gender','Medal']]       #creating a subset with only the required columns
    seperated_data=seperated_data[seperated_data['Gender']==gender]       #keeping the data based on the gender specified
    Highest_medals=seperated_data.Athlete.value_counts().sort_values(ascending=False)   #WIll count the number of times athlete name appeared as rw ressembles a medal won
    return Highest_medals.head(1) #Returning the first row
```

```python
highest=calculateNumberofMedals(summer,"Men")
print(highest)
```

```python
highest=calculateNumberofMedals(summer,"Women")
print(highest.head(1))
```

7)	Create a function to find the athlete who won the highest number of medals in each medal category (Gold, Bronze, Silver). 

```python
def calculateNumberofMedalsinCategory(dataset):
    copy_data=dataset.copy()         #Create a copy of dataframe
    seperated_data=copy_data[['Athlete','Medal']]       #Filtering the required columns
    seperated_data_gold=seperated_data[seperated_data['Medal']=='Gold']      #Creating a subset where only gold medal data is kept
    seperated_data_silver=seperated_data[seperated_data['Medal']=='Silver'] #Creating a subset where only silver medal data is kept
    seperated_data_bronze=seperated_data[seperated_data['Medal']=='Bronze'] #Creating a subset where only bronze medal data is kept
    Gold=seperated_data_gold.Athlete.value_counts().sort_values(ascending=False).head(1)   #Getting the athlete with highest gold medals
    Silver=seperated_data_silver.Athlete.value_counts().sort_values(ascending=False).head(1) #Getting the athlete with highest silver medals
    Bronze=seperated_data_bronze.Athlete.value_counts().sort_values(ascending=False).head(1) #Getting the athlete with highest bronze medals
    df = pd.concat([Gold, Silver, Bronze]).to_frame()    #concatination of the three dataframe in to a single frame
    df['Medal type'] = ['Gold','Silver','Bronze']   
    df=df.rename({'Athlete':'No. of medals'},axis=1)
    return df
```

```python
category_count= calculateNumberofMedalsinCategory(winter)
print(category_count)
```
